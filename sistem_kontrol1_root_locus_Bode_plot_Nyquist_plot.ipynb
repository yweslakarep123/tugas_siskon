```python
# nomor 1A
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

num1 = [1, 5, 6]
den1 = [1, 0,-1]

num2 = [1]
den2 = [1,15,50,0]

num3 = [1]
den3 = [1,6,11,6]

W1 = ctrl.tf(num1,den1)
print(W1)

W2 = ctrl.tf(num2,den2)
print(W2)

W3 = ctrl.tf(num3,den3)
print(W3)

xAxisLimits = (-13.0, 0.5)
yAxisLimits = (-2.5 , 2.5)

roots, gains = ctrl.root_locus(W1, xlim=xAxisLimits, ylim=yAxisLimits, plot=True)


print(W1.poles())
print(W2.poles())
print(W3.poles())

plt.grid()
plt.show()
```

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

num1 = [1, 5, 6]
den1 = [1, 0,-1]

num2 = [1]
den2 = [1,15,50,0]

num3 = [1]
den3 = [1,6,11,6]


W1 = ctrl.tf(num1,den1)
print(W1)

W2 = ctrl.tf(num2,den2)
print(W2)

W3 = ctrl.tf(num3,den3)
print(W3)

count1, contour1 = ctrl.nyquist_plot(W1, return_contour=True, arrows = 2,
                                   arrow_size=12, start_marker='o', start_marker_size=12, linewidth=4, plot=True)

plt.show()
```
